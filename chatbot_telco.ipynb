```python
import os
import re
from pathlib import Path
import pdfplumber

def read_pdf(file_path):
    text = ""
    # Open the PDF file
    with pdfplumber.open(file_path) as pdf:
        # Iterate through all the pages and extract text
        for page in pdf.pages:
            text += page.extract_text()
    return text

def clean_text(text):
    text = re.sub(r'[^\x20-\x7E]+', ' ', text)

    text = re.sub(r'\s', ' ', text)

    text = text.strip()

    return text


# Determine the current directory based on the environment
if '__file__' in globals():
    # Running as a script
    current_dir = Path(__file__).parent
else:
    # Running in an interactive environment
    current_dir = Path(os.getcwd())

# Specify the relative path to the PDF file
pdf_file_path = current_dir / 'Design_and_Analysis.pdf'

# Check if the file exists
if not pdf_file_path.exists():
    print(f"The file {pdf_file_path} does not exist.")
else:
    # Read the PDF
    pdf_text = read_pdf(pdf_file_path)
    cleaned_text = clean_text(pdf_text)
    print(cleaned_text)
```

```python
from transformers import BertTokenizer, BertModel, TFBertForSequenceClassification
import torch

# Load pre-trained model tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_sentence_embedding(sentence):
    inputs = tokenizer(sentence, return_tensors='pt', truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1)

# Generate embeddings for each sentence
embeddings = [get_sentence_embedding(sentence) for sentence in sentences]
```

```python
from sklearn.cluster import KMeans

# Convert list of embeddings to a numpy array
embeddings_np = torch.cat(embeddings).numpy()

# Perform K-means clustering
OMP_NUM_THREADS=2
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(embeddings_np)

# Assign labels to each sentence
labels = kmeans.labels_
print(labels)
```

```python
# Example mapping of cluster labels to human-readable labels
cluster_to_label = {
    0: 'Introduction',
    1: 'Methods',
    2: 'Results',
    3: 'Discussion',
    4: 'Conclusion'
}

labeled_sentences = [(sentence, cluster_to_label[label]) for sentence, label in zip(sentences, labels)]

for sentence, label in labeled_sentences:
    print(f"Sentence: {sentence}\nLabel: {label}\n")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming df is the DataFrame created previously
# Split the data into training and validation sets
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Separate features and labels
train_texts, train_labels = train_df['Sentence'].tolist(), train_df['Label'].tolist()
val_texts, val_labels = val_df['Sentence'].tolist(), val_df['Label'].tolist()
```

```python
# Encode labels as integers
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_labels)
val_labels = label_encoder.transform(val_labels)
```

```python
from transformers import BertTokenizer
import tensorflow as tf
import torch

# Load pre-trained model tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize and encode the text data
def encode_data(texts, labels, tokenizer, max_length=128):
    input_ids = []
    attention_masks = []

    for text in texts:
        encoded = tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=max_length,
            pad_to_max_length=True,
            return_attention_mask=True,
            return_tensors='pt',
        )
        input_ids.append(encoded['input_ids'])
        attention_masks.append(encoded['attention_mask'])

    input_ids = tf.concat(input_ids, axis=0)
    attention_masks = tf.concat(attention_masks, axis=0)
    labels = tf.convert_to_tensor(labels)
    return input_ids, attention_masks, labels

train_inputs, train_masks, train_labels = encode_data(train_texts, train_labels, tokenizer)
val_inputs, val_masks, val_labels = encode_data(val_texts, val_labels, tokenizer)
```

```python
# Convert to tf.data.Dataset
def create_dataset(inputs, masks, labels, batch_size=32):
    dataset = tf.data.Dataset.from_tensor_slices(({'input_ids': inputs, 'attention_mask': masks}, labels))
    dataset = dataset.shuffle(buffer_size=10000).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
    return dataset
```

```python
train_dataset = create_dataset(train_inputs, train_masks, train_labels)
```

```python
val_dataset = create_dataset(val_inputs, val_masks, val_labels, batch_size=32)
```

```python
# Load pre-trained BERT model for sequence classification
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(cluster_to_label))
```

```python
from tensorflow.keras.optimizers import SGD
import tensorflow as tf
os.environ['TF_USE_LEGACY_KERAS'] = '1'
from sklearn.metrics import classification_report
# Compile the model
optimizer = optimizer = tf.keras.optimizers.Adam()
model.compile(optimizer='SGD', loss=model.compute_loss, metrics=['accuracy'])
```

```python
history = model.fit(
    [train_inputs, train_masks],
    train_labels,
    validation_data=([val_inputs, val_masks], val_labels),
    epochs=3,
     batch_size=32
    )
```

```python
import numpy as np
val_loss, val_accuracy = model.evaluate([val_inputs, val_masks], val_labels, batch_size=32)
print(f'Validation loss: {val_loss}')
print(f'Validation accuracy: {val_accuracy}')
        
# Make predictions
y_pred = model.predict([val_inputs, val_masks]).logits
y_pred = np.argmax(y_pred, axis=1)
        
# Print classification report
print(classification_report(val_labels, y_pred, target_names=cluster_to_label.values()))
        
# Save the model
model.save_pretrained('fine_tuned_bert_model')
tokenizer.save_pretrained('fine_tuned_bert_tokenizer')
```
